```python
# importing libraries, loading data and validation
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("DA -Task 2..xlsx")
```

```python
df.head()
df.tail()
df.shape
df.size
df.columns
```

```python
# Fixing cases and removing spaces from column
df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")
```

```python
# column wise structure analysis
df.info()
df.describe(include="all") #inculded all, so even non-numeric can be analysed
df.nunique().sort_values(ascending=False)
```

```python
# dealing with missing values
# Finding:-
null_summary = df.isna().sum().sort_values(ascending=False)
null_percentage = (df.isna().mean()*100).sort_values(ascending=False)
null_summary, null_percentage # Got to know what columns do have missing values and it's percentage
```

```python
# Dealing with the findings:-
df.drop(columns=['campaign_nbr'], inplace=True) # Had 100% null rate
```

```python
cat_cols = ['engine_source_plant','transmission_trace_nbr', 'engine_trace_nbr', 'optn_famly_certification','optf_famly_emissiof_system',
            'causal_part_nm', 'repair_dlr_postal_cd', 'veh_test_grp', 'state', 'line_series', 'plant']

for col in cat_cols:
    df[col] = df[col].fillna("Unknown")
```

```python
num_cols = ['transmission_source_plant', 'totalcost', 'last_known_delvry_type_cd']

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
# Duplicates - finding and dropping
df.duplicated().sum() # Got to know no duplicates, so nothing dropped
```

```python
#Data type validation
df['repair_date'] = pd.to_datetime(df['repair_date'], errors='coerce')
df['repair_date'].isna().sum() # to ensure the date column is valid and ready for analysis
df.info()
```

```python
#Tag creation - created tagging based on text
def generate_tag(text):
    if pd.isna(text):
        return "unknown"
    text = text.lower()
    if "noise" in text:
        return "noise_issue"
    elif "leak" in text:
        return "leakage"
    elif "steering" in text:
        return "steering_issue"
    elif "overheat" in text:
        return "overheating"
    else:
        return "other"

df['failure_tag'] = df['customer_verbatim'].apply(generate_tag)
```

```python
#Visualization
df['causal_part_nm'].value_counts().head(10).plot(kind='bar')
plt.title("Top Failed Components")
plt.xlabel("Component")
plt.ylabel("Count")
plt.show()
```

```python
df.groupby('platform')['totalcost'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title("Total Warranty Cost by Platform")
plt.ylabel("Total Cost")
plt.show()
```

```python
df['global_labor_code_description'].value_counts().head(10).plot(kind='bar')
plt.title("Most Frequent Repair Actions")
plt.show()
```

```python
df['failure_tag'].value_counts().head(3).plot(kind='bar')
plt.title('Top Failures')
plt.xlabel('Failures')
plt.ylabel('Count')
plt.show()
```
